## merging all tables for further analysis

```python
merged_data = pd.merge(pizza_sales_df,pizza_category_df, on = 'pizza_category_id')
merged_data
```

```python
merged_data = pd.merge(merged_data,pizza_size_df, on = 'pizza_size_id')
merged_data
```

```python
merged_data = pd.merge(merged_data,pizza_sales_voucher_df, on = 'order_details_id')
merged_data
```

### Now we had one dataframe has all data called 'merged_data' and the column names are as the following:

```python
merged_data.columns
```

```python
merged_data['Month_Year'] = merged_data['order_date'].dt.to_period('M')
merged_data['Month_Year']
```
